```python
# convert data using converter - each data check of Temperature column and change the data
def convert_data(data):
    if data=="38":
        return '39'
    return data
df = pd.read_csv("weather.csv", converters={
    'Temperature' : convert_data
})
print(df.head())
```

```python
# replace 
new_df = df.replace(3, np.NaN)
new_df.head()
```

```python
# replace multiple value using list
new_df = df.replace([23, 21],np.NaN)
new_df.head()
```

```python
# replacement based on columns
new_df = df.replace({
    'DewPoint': 18,
    'Humidity': 46
}, np.NaN)
print(new_df.head())
```

```python
# replace - mapping
new_df = df.replace({
    18 : np.NaN,
    46 : np.NaN
})
print(new_df.head())
```
